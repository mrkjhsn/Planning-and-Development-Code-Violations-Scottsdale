### Using the latitude and longitude coordinates for each code violation I wanted to explore areas of the city that some code violations may be clustered in. 

```python
# extra libraries for gis plotting
import geoviews as gv
import hvplot.pandas
import hvplot.xarray
import geopandas as gpd

# standard libraries
import numpy as np
import matplotlib.pyplot as plt
import pyodbc
import pandas as pd
import seaborn as sns
pd.set_option('display.max_columns', None)

# databaseConnection = 'DRIVER={SQL Server}; SERVER=ServerName; Database=DatbaseName; UID=UserId; PWD=password;'
connection = pyodbc.connect('DRIVER={SQL Server};SERVER=WINDOWS-A6S01I2\SQLEXPRESS;DATABASE=City_of_Scottsdale;Trusted_Connection=yes;')
cursor = connection.cursor()

# SQL query to import all citation data
code_violations_query = ''' select * from [City_of_Scottsdale].[dbo].[np_CodeViolations] '''

code_violations = pd.read_sql(code_violations_query, connection)
connection.close()

# convert latitude and longitude attributes to numeric from object
# can I combine both of the below lines of code into one?
code_violations['_Latitude_'] = pd.to_numeric(code_violations['_Latitude_'], errors="coerce")
code_violations['_Longitude_'] = pd.to_numeric(code_violations['_Longitude_'], errors="coerce")

# set 'ViolationCategory' as index
code_violations.set_index('_ViolationCategory_', inplace=True)

code_violations.head()
```

### Basic visualization of the location of code violations

```python
plt.rcParams["figure.figsize"] = [15, 20]

plt.scatter(x=code_violations['_Longitude_'], y=code_violations['_Latitude_'])
plt.ylabel("Longitude")
plt.xlabel("Latitude")

# plot the mean latitude and longitude coordinates for all violations
plt.axhline(code_violations['_Latitude_'].mean(), color='grey', label='mean violation', linestyle='dashed')
plt.axvline(code_violations['_Longitude_'].mean(), color='grey', label='mean violation', linestyle='dashed')
```

## Clustering analysis
### Which code violation categories have a mean location furthest from the mean location of all code violations?

```python
code_violations.head()
```

```python
# limit to groups with at least 20 violations over the course of the year
# so many violations at either end of the spectrum(north/south) have only a handful of occurences

over_twenty = code_violations.groupby('_ViolationCategory_') \
    .filter(lambda x: x['_Latitude_'].count() >= 10)

# after filtering, 30 unique violations remain
over_twenty.index.unique().to_frame().count()
```

```python
over_twenty_count = over_twenty.groupby('_ViolationCategory_')['_ViolationID_'].count().to_frame()
over_twenty_count.head()
```

```python
over_twenty_grouped = over_twenty.groupby('_ViolationCategory_')['_Latitude_' , '_Longitude_'] \
    .agg(['mean'])

over_twenty_grouped.head()
```

```python
# create columns for latitude mean and longitude mean
over_twenty_grouped['mean_lat'] = code_violations['_Latitude_'].mean()
over_twenty_grouped['mean_lon'] = code_violations['_Longitude_'].mean()

over_twenty_grouped.rename(columns={'_Latitude_': 'mean_group_lat',
                                  '_Longitude_': 'mean_group_lon',
                                  }, inplace=True)

# flatten column labels into one level
over_twenty_grouped.columns = over_twenty_grouped.columns.get_level_values(0)

over_twenty_grouped.head()
```

```python
# find the difference from the mean for each code violation grouping
over_twenty_grouped['dif_from_mean'] = np.sqrt(np.square(abs(over_twenty_grouped.iloc[:,2]) - \
                                        abs(over_twenty_grouped.iloc[:,0])) \
                                        + \
                                        np.square(abs(over_twenty_grouped.iloc[:,3]) - \
                                        abs(over_twenty_grouped.iloc[:,1]))) 
```

```python
top_violations = over_twenty_grouped.sort_values(by='dif_from_mean', ascending=False).head(12)
```

```python
top_violations = top_violations.merge(over_twenty_count,  left_index=True, right_index=True)

top_violations
```

```python
fig = plt.figure(figsize=(15, 20))

sns.set_context("notebook", font_scale=1)


# plot all violation points
ax = sns.scatterplot(x=code_violations['_Longitude_'], y=code_violations['_Latitude_'], color='skyblue', s=20)

# plot top violation points, scaled by quantity of those violations
ax = sns.scatterplot(x=top_violations.iloc[:,1], 
            y=top_violations.iloc[:,0],
            hue=top_violations.index,
            s=top_violations.iloc[:,5]*100,
            alpha=0.5)

# plot the mean latitude and longitude coordinates for all violations
plt.axhline(code_violations['_Latitude_'].mean(), color='grey', label='mean violation', linestyle='dashed')
plt.axvline(code_violations['_Longitude_'].mean(), color='grey', label='mean violation', linestyle='dashed')

plt.ylabel("Latitude", fontsize=15)
plt.xlabel("Longitude", fontsize=15)

# plt.xlabel("Day of the Year", fontsize=15)
# plt.ylabel("Fund", fontsize=15)
# plt.savefig("visualizations/violation categories farthest from mean violation category.png")
```
